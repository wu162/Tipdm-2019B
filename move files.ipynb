```python
#这个是将全部数据里的dcm和mask归类的
import os
import shutil
INPUT_FOLDER = './B-allData/data/'
patients = os.listdir(INPUT_FOLDER)
patients.sort()
AP_FLODER = '/arterial phase/'
VP_FLODER = '/venous phase/'
```

```python
for path in patients:
    print("start copy", path)
    old_ap_path = INPUT_FOLDER + path + AP_FLODER
    old_vp_path = INPUT_FOLDER + path + VP_FLODER
    new_ap_path ='./B-allData/apDCM/'+path+'/'
    new_mask_path ='./B-allData/mask/'+path+'/'
    os.mkdir(new_ap_path)
    os.mkdir(new_mask_path)
    for filename in os.listdir(old_ap_path):
        if 'dcm' in filename:
            shutil.copy(old_ap_path + filename, new_ap_path+filename)
        else:
            shutil.copy(old_ap_path + filename, new_mask_path+filename)
```
